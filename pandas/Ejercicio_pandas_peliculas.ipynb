## Cálculos, manipulación y visualización

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Crear variable Rentabilidad

df_peliculas['Rentabilidad'] = round(df_peliculas['Ganancias'] / df_peliculas['Presupuesto'] * 100, 2)
df_peliculas
```

```python
# Buscar relaciones entre género, rentabilidad y duración y mostrar gráfico de línea
df_relacion_genero_rentabilidad_duracion = df_peliculas.groupby('Género')[['Rentabilidad', 'Duración (minutos)']].mean()

# Crear un gráfico de líneas
fig, ax1 = plt.subplots(figsize=(12, 8))

# Primer eje y (Rentabilidad)
color = 'tab:red'
ax1.set_xlabel('Género')
ax1.set_ylabel('Rentabilidad', color=color)
ax1.plot(df_relacion_genero_rentabilidad_duracion.index, df_relacion_genero_rentabilidad_duracion['Rentabilidad'], color=color, marker='o')
ax1.tick_params(axis='y', labelcolor=color)

# Segundo eje y (Duración)
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Duración (minutos)', color=color)
ax2.plot(df_relacion_genero_rentabilidad_duracion.index, df_relacion_genero_rentabilidad_duracion['Duración (minutos)'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

# Mostrar el gráfico
plt.title('Relación entre Género, Rentabilidad y Duración')
plt.show()
```

```python
# Buscar relaciones entre el presupuesto y la calificación IMDB por género
df_relacion_rentabilidad_calificacion_imdb = df_peliculas.groupby('Género')[['Presupuesto', 'Calificación IMDB']].mean()
```

```python
# Crear el gráfico para mostrar la relación entre presupuesto, género y calificación IMBd

# Crear una figura y ejes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Graficar barras para el presupuesto
color = 'tab:blue'
ax1.set_xlabel('Género')
ax1.set_ylabel('Presupuesto', color=color)
bar_plot = ax1.bar(df_relacion_rentabilidad_calificacion_imdb.index, df_relacion_rentabilidad_calificacion_imdb['Presupuesto'], color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Crear ejes secundarios para la calificación IMDb
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Calificación IMDb', color=color)
line_plot = ax2.plot(df_relacion_rentabilidad_calificacion_imdb.index, df_relacion_rentabilidad_calificacion_imdb['Calificación IMDB'], color=color, marker='o')
ax2.tick_params(axis='y', labelcolor=color)

# Añadir etiquetas y título
plt.title('Relación entre Presupuesto y Calificación IMDb por Género')
plt.show()
```
