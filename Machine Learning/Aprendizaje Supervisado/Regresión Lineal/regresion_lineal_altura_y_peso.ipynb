# Ejercicio: Predecir el peso en función de la altura

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Lectura del conjunto de datos
df_altura_peso = pd.read_csv('Data EJERCICIO_1 Regresión Lineal_Con Test.csv', delimiter=';')
df_altura_peso
```

```python
# División del conjunto de datos en características (altura) y variable objetivo (peso)
X = df_altura_peso[['Altura']]
y = df_altura_peso[['Peso']]
```
