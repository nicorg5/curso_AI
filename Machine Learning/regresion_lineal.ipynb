```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# ¿Cuánto coste se tendría si se afectan 1.830 equipos?

X = 1000 * 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.abs(1000 * (np.random.randn(100, 1))) #se añade .abs para que no haya valores negativos
```

```python
# Se crea y se entrena el modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)
```

```python
# Realizar la predicción para 1,830 equipos
equipos_nuevos = np.array([[1830]])
costo_predicho = model.predict(equipos_nuevos)
```

```python
print(f"Costo estimado para 1830 equipos: {costo_predicho[0][0]}")
```

```python
# Crear un diagrama de dispersión con la línea de regresión y el valor predicho
plt.scatter(X, y, label='Datos')
plt.plot(X, model.predict(X), color='red', label='Línea de regresión')
plt.scatter(equipos_nuevos, costo_predicho, color='green', marker='X', s=100, label='1,830 equipos predichos')

# Configurar el gráfico
plt.title('Regresión Lineal - Equipos vs. Costo')
plt.xlabel('Equipos afectados')
plt.ylabel('Costo')
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
