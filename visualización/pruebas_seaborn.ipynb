```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

# Conjunto de datos 'exercise'

```python
# Cargar el conjunto de datos "exercise"
exercise = sns.load_dataset("exercise")
exercise
```

```python
# Histograma del pulso con líneas de distribución

sns.histplot(exercise["pulse"], kde=True)
```

```python
# Gráfico de barras para ver la relación entre el tiempo y el pulso según el tipo de ejercicio

sns.barplot(x="time", y="pulse", data=exercise, hue="kind")
```

# Taxis

```python
# Cargar el conjunto de datos "taxis"
taxis = sns.load_dataset("taxis")
taxis
```

```python
# Grafico de dispersión entre distancia y tarifa en función de los pasajeros

sns.scatterplot(x="distance", y="fare", data=taxis, hue="passengers")
plt.title('Relación entre Distancia y Tarifa')
```

```python
# Histograma con líneas de distribución (kde) para la distancia

sns.histplot(taxis["distance"], kde=True)
plt.title('Distribución de Distancias')
```

```python
# Gráfico para ver la distancia en función del método de pago

distance_by_zone = taxis.groupby('payment')['distance'].mean().reset_index()
sns.barplot(x='payment', y='distance', data=distance_by_zone, palette='viridis')
```
