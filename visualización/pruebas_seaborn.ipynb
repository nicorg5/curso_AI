```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

# Conjunto de datos 'exercise'

```python
# Cargar el conjunto de datos "exercise"
exercise = sns.load_dataset("exercise")
exercise
```

```python
#Â Histograma del pulso con lÃ­neas de distribuciÃ³n

sns.histplot(exercise["pulse"], kde=True)
```

```python
# GrÃ¡fico de barras para ver la relaciÃ³n entre el tiempo y el pulso segÃºn el tipo de ejercicio

sns.barplot(x="time", y="pulse", data=exercise, hue="kind")
```

# Taxis

```python
# Cargar el conjunto de datos "taxis"
taxis = sns.load_dataset("taxis")
taxis
```

```python
# Grafico de dispersiÃ³n entre distancia y tarifa en funciÃ³n de los pasajeros

sns.scatterplot(x="distance", y="fare", data=taxis, hue="passengers")
plt.title('RelaciÃ³n entre Distancia y Tarifa')
```

```python
# Histograma con lÃ­neas de distribuciÃ³n (kde) para la distancia

sns.histplot(taxis["distance"], kde=True)
plt.title('DistribuciÃ³n de Distancias')
```

```python
# GrÃ¡fico para ver la distancia en funciÃ³n del mÃ©todo de pago

distance_by_zone = taxis.groupby('payment')['distance'].mean().reset_index()
sns.barplot(x='payment', y='distance', data=distance_by_zone, palette='viridis')
```
