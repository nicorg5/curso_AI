```python
from astrapy.db import AstraDB
import pandas as pd
import os
import subprocess
from cassandra.cluster import Cluster
from dotenv import load_dotenv
```

```python
# Cargar variables de entorno
load_dotenv()
# Configurar el token de AstraDB
astradb_token=os.getenv("ASTRADB_TOKEN")
```

```python
# Conexión a la base de datos
db = AstraDB(
  token=astradb_token,
  api_endpoint="https://f1ef8e9b-dc05-412a-86a9-4712d299b57d-us-east-1.apps.astra.datastax.com")

print(f"Connected to Astra DB: {db.get_collections()}")
```

```python
# Buscar una colección
collection = db.collection("movie_reviews")
collection
```

```python
# Buscar documentos dentro de la colección
results = collection.find()
results
```

```python
# 'results' es un diccionario en el que está la información que interesa convertir en un DataFrame
documents = results['data']['documents']
print(f"La variable documents es una {type(documents)} y tiene la siguiente información: \n{documents}")
```

```python
# Se crea un DataFrame
df_movie_reviews = pd.DataFrame(documents)
df_movie_reviews
```
